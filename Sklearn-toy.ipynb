```python
import os
import pandas as pd
```

```python
path = os.getcwd().replace("\\", "/")
data = pd.read_csv(path + "/toy_dataset.csv")
del data['Number']
```

```python
list_of_city = []
for city in data['City']:
    if city not in list_of_city:
        list_of_city.append(city)
```

```python
city = pd.get_dummies(data['City'])
gender = pd.get_dummies(data['Gender'])
data = data.merge(city, left_index=True, right_index=True)
data = data.merge(gender, right_index = True, left_index = True)
data
```

```python
data["target"] = data["Illness"]
del data["Illness"]
del data["City"]
del data["Gender"]
data
```

```python
from sklearn.preprocessing import StandardScaler
std_scl = StandardScaler()
data['Age'] = std_scl.fit_transform(data['Age'])
data['Income'] = std_scl.fit_transform(data['Income'])
```

```python
features = data.iloc[:,0:-1]
target = data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
knn = KNeighborsClassifier(3)
dtree = DecisionTreeClassifier()
svm = SVC(kernel="rbf")
nn = MLPClassifier(hidden_layer_sizes=(5), activation='relu', solver='adam')
```

```python
knn.fit(x_train, y_train)
dtree.fit(x_train, y_train)
svm.fit(x_train, y_train)
nn.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
knn_pred = knn.predict(x_test)
dtree_pred = dtree.predict(x_test)
svm_pred = svm.predict(x_test)
nn_pred = nn.predict(x_test)
```

```python
print(confusion_matrix(y_test, knn_pred))
print(confusion_matrix(y_test, dtree_pred))
print(confusion_matrix(y_test, svm_pred))
print(confusion_matrix(y_test, nn_pred))
print(classification_report(y_test, knn_pred))
print(classification_report(y_test, dtree_pred))
print(classification_report(y_test, svm_pred))
print(classification_report(y_test, nn_pred))
```
