```python
import pandas as pd
```

```python
data = pd.read_csv("D:/Code/dataset/Iris.csv")
data
```

```python
data.columns
```

```python
features = data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
label = data['Species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, label, test_size= 0.25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(5)
clf.fit(x_train, y_train)
```

```python
pred1 = clf.predict(x_test)
```

```python
from sklearn.svm import SVC
clf2 = SVC(kernel='linear')
clf2.fit(x_train, y_train)
```

```python
pred2 = clf2.predict(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf3 = DecisionTreeClassifier()
clf3.fit(x_train, y_train)
```

```python
pred3 = clf3.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(pred1, y_test))
print(classification_report(pred2, y_test))
print(classification_report(pred3, y_test))
```

```python
print(confusion_matrix(pred1, y_test))
print(confusion_matrix(pred2, y_test))
print(confusion_matrix(pred3, y_test))
```
