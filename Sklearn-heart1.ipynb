```python
import pandas as pd
import os
```

```python
path = os.getcwd().replace("\\", "/")
```

```python
data = pd.read_csv(path + "/heart.csv")
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scl = StandardScaler()
```

```python
data['age'] = std_scl.fit_transform(data['age'])
data['trtbps'] = std_scl.fit_transform(data['trtbps'])
data['chol'] = std_scl.fit_transform(data['chol'])
data['thalachh'] = std_scl.fit_transform(data['thalachh'])
data['oldpeak'] = std_scl.fit_transform(data['oldpeak'])
data['thall'] = std_scl.fit_transform(data['thall'])
```

```python
features = data.iloc[:,0:-1]
target = data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size= 0.25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
```

```python
knn = KNeighborsClassifier(3)
knn.fit(x_train, y_train)
pred1 = knn.predict(x_test)
```

```python
svm = SVC(kernel = 'rbf')
svm.fit(x_train, y_train)
pred2 = svm.predict(x_test)
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
pred3 = dtree.predict(x_test)
```

```python
nn = MLPClassifier(hidden_layer_sizes = (3), activation = 'relu', solver = 'adam')
nn.fit(x_train, y_train)
pred4 = nn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test, pred1))
print(confusion_matrix(y_test, pred2))
print(confusion_matrix(y_test, pred3))
print(confusion_matrix(y_test, pred4))
print('KNN:')
print(classification_report(y_test, pred1))
print('SVC:')
print(classification_report(y_test, pred2))
print('DTree:')
print(classification_report(y_test, pred3))
print('Neural Network:')
print(classification_report(y_test, pred4))
```
