```python
import os
import pandas as pd
```

```python
path = os.getcwd()
path.replace("\\", "/")
```

```python
data = pd.read_csv(path + "/heart.csv")
data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scl = StandardScaler()
data['age'] = std_scl.fit_transform(data['age'])
data['cp'] = std_scl.fit_transform(data['cp'])
data['trtbps'] = std_scl.fit_transform(data['trtbps'])
data['chol'] = std_scl.fit_transform(data['chol'])
data['thalachh'] = std_scl.fit_transform(data['thalachh'])
data['oldpeak'] = std_scl.fit_transform(data['oldpeak'])
data['thall'] = std_scl.fit_transform(data['thall'])
```

```python
data
```

```python
features = data.iloc[:,0:-1]
target = data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size= 0.25)
```

```python
from sklearn.neural_network import MLPRegressor
mlp = MLPRegressor(hidden_layer_sizes = (128,128,128), activation = "relu", random_state = 1, max_iter = 2000)
mlp.fit(x_train, y_train)
```

```python
parameter_space = {
    'hidden_layer_sizes' : {128,128,128},
    'activation' : ['tanh', 'relu'],
    'solver' : ['sgd', 'adam'],
    'alpha' : [0.0001, 0.05],
    'learning_rate' : ['constant', 'adaptive'],
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(parameter_space)
clf.fit(x_train, y_train)
print('Best parameters found:\n', clf.best_parameter)
```
